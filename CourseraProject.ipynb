```python
#import the librarys

import requests
from bs4 import BeautifulSoup
import csv
from time import sleep
import pandas as pd

url = 'https://www.randstad.com/jobs/canada/s-administrative-support-services/'

r = requests.get(url)

soup = BeautifulSoup(r.text, 'lxml')
```

```python
soup
```

```python
# get the link 

link = 'https://www.randstad.com'+soup.find('li', class_= 'cards__item bg-variant-white').find('a', class_= 'cards__link' ).get('href')

link
```

```python
# get the job title

job_title = soup.find('li', class_= 'cards__item bg-variant-white').find('a', class_= 'cards__link').text

job_title
```

```python
# get the location of role

location = soup.find('li', class_ = 'cards__item bg-variant-white').findAll('li', class_ = 'cards__meta-item' )[0].text
location
```

```python
# get the job type 

job_type = soup.find('li', class_ = 'cards__item bg-variant-white').findAll('li', class_ = 'cards__meta-item')[1].text

job_type
```

```python
# get the renumeration indication of role

renumeration = soup.find('li', class_ = 'cards__item bg-variant-white').findAll('li', class_ = 'cards__meta-item')[2].text

renumeration
```

```python
#  get the job  description

job_description = soup.find('li', class_ = 'cards__item bg-variant-white').find('div', class_ = 'cards__description').text

job_description
```

```python
# get the publication date

published = soup.find('span', class_ = 'cards__date text--alternative').text

published
```

```python
# get all jobs
data = []

for p in range(1,42):
    print(p)
    
    url = f"https://www.randstad.com/jobs/canada/s-administrative-support-services/page-{p}/"
    r = requests.get(url)
    sleep(3)
    soup = BeautifulSoup(r.text, 'lxml')



    jobs = soup.findAll('li', class_= 'cards__item bg-variant-white')



    for job in jobs:
        link = 'https://www.randstad.com'+job.find('a', class_= 'cards__link' ).get('href')
        job_title = job.find('a', class_= 'cards__link').text
        location = job.findAll('li', class_ = 'cards__meta-item' )[0].text
        try:
            job_type = job.findAll('li', class_ = 'cards__meta-item')[1].text
        except:
            job_type = '-'
        try:
            renumeration = job.findAll('li', class_ = 'cards__meta-item')[2].text
        except:
            renumeration = '-'
        try:
            job_description = job.find('div', class_ = 'cards__description').text
        except:
            job_description = '-'
        try:
            published = job.find('span', class_ = 'cards__date text--alternative').text
        except:
            published = '-'
        data.append([link, job_title, location, job_type, renumeration, job_description, published ])
 
```

```python
len(data)
```

```python
# column names of DataFrame

header = ['link', 'job_title', 'location', 'job_type', 'renumeration', 'job_description', 'published']
```

```python
# make DataFrame

df = pd.DataFrame(data, columns=header)
df.to_csv('jobs_administrative-support-services.csv', sep = ',', encoding = 'utf8')
```
